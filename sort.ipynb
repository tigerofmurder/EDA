```python
def bubbleSort(arr): 
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j]
                
"""A = list(range(1,10000))
shuffle(A)
bubbleSort(A)
print(A)
"""
```

```python
import random, math

def countingSort(tlist, k):
    count_list = [0]*(k)
    for n in tlist:
        count_list[n] = count_list[n] + 1
    i=0
    for n in range(len(count_list)):
        while count_list[n] > 0:
            tlist[i] = n
            i+=1
            count_list[n] -= 1
            
def countSort(arr):
    countingSort(arr, max(arr)+1)
```

```python
def heapify(arr, n, i): 
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < n and arr[i] < arr[l]: 
        largest = l
    if r < n and arr[largest] < arr[r]: 
        largest = r
    if largest != i: 
        arr[i],arr[largest] = arr[largest],arr[i]
        heapify(arr, n, largest) 

def heapSort(arr): 
    n = len(arr) 
    for i in range(n, -1, -1): 
        heapify(arr, n, i)
    for i in range(n-1, 0, -1): 
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0) 
```

```python
def insertionSort(arr): 
    for i in range(1, len(arr)): 
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j] : 
                arr[j + 1] = arr[j] 
                j -= 1
        arr[j + 1] = key 
```

```python
def mergeSort(arr): 
    if len(arr) >1: 
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
  
        mergeSort(L)
        mergeSort(R)
  
        i = j = k = 0
        while i < len(L) and j < len(R): 
            if L[i] < R[j]: 
                arr[k] = L[i] 
                i+=1
            else: 
                arr[k] = R[j] 
                j+=1
            k+=1
        while i < len(L): 
            arr[k] = L[i] 
            i+=1
            k+=1
          
        while j < len(R): 
            arr[k] = R[j] 
            j+=1
            k+=1
```

```python
def partition(arr,low,high): 
    i = ( low-1 )
    pivot = arr[high]
    for j in range(low , high):
        if   arr[j] < pivot:
            i = i+1 
            arr[i],arr[j] = arr[j],arr[i] 
    arr[i+1],arr[high] = arr[high],arr[i+1] 
    return ( i+1 ) 
  
def iquickSort(arr,low,high): 
    if low < high: 
        pi = partition(arr,low,high)
        iquickSort(arr, low, pi-1) 
        iquickSort(arr, pi+1, high) 

def quickSort(arr):
    n = len(arr) 
    iquickSort(arr,0,n-1) 
```

```python
def selectionSort(A):
    for i in range(len(A)): 
        min_idx = i 
        for j in range(i+1, len(A)): 
            if A[min_idx] > A[j]: 
                min_idx = j 
        A[i], A[min_idx] = A[min_idx], A[i]
```

```python
def tryItABunch(myFn, startN=10, endN=100, stepSize=10, numTrials=20, listMax = 10):
    nValues = []
    tValues = []
    for n in range(startN, endN, stepSize):
        # run myFn several times and average to get a decent idea.
        runtime = 0
        for t in range(numTrials):
            lst = [ choice(range(listMax)) for i in range(n) ] # generate a random list of length n
            start = time.time()
            myFn( lst )
            end = time.time()
            runtime += (end - start) * 1000 # measure in milliseconds
        runtime = runtime/numTrials
        nValues.append(n)
        tValues.append(runtime)
    return nValues, tValues
```

```python
nValuesNaive, tValuesNaive = tryItABunch( bubbleSort, startN = 50, endN = 10000, stepSize=500, numTrials=1, listMax = 500 )

nValues, tValues = tryItABunch( insertionSort,startN = 50, endN = 10000, stepSize=500, numTrials=1, listMax = 500 )#

nValuesC, tValuesC = tryItABunch( countSort, startN = 50, endN = 10000, stepSize=500, numTrials=1, listMax = 500 )

nValuesH, tValuesH = tryItABunch( heapSort, startN = 50, endN = 10000, stepSize=500, numTrials=1, listMax = 500 )

nValuesM, tValuesM = tryItABunch( mergeSort, startN = 50, endN = 10000, stepSize=500, numTrials=1, listMax = 500 )

nValuesQ, tValuesQ = tryItABunch( quickSort, startN = 50, endN = 10000, stepSize=500, numTrials=1, listMax = 500 )

nValuesS, tValuesS = tryItABunch( selectionSort, startN = 50, endN = 10000, stepSize=500, numTrials=1, listMax = 500 )
```

```python
plt.plot(nValuesNaive, tValuesNaive, color="red", label="BubbleSort")
plt.plot(nValues, tValues, color="blue", label="InsertionSort")
plt.plot(nValuesC, tValuesC, color="green", label="CountingSort")
plt.plot(nValuesH, tValuesH, color="black", label="HeapSort")
plt.plot(nValuesM, tValuesM, color="skyblue", label="MergeSort")
plt.plot(nValuesQ, tValuesQ, color="pink", label="QuickSort")
plt.plot(nValuesS, tValuesS, color="brown", label="SelectionSort")
plt.xlabel("n")
plt.ylabel("Time(ms)")
plt.ylim(top=100) 
plt.legend()
```
