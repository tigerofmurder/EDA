```python
import numpy as np
import matplotlib.pyplot as plt


x4,y4 = np.loadtxt('mergesecond.txt',delimiter=' ',unpack=True)
sx4,sy4 = np.loadtxt('mergefirst.txt',delimiter=' ',unpack=True)
zx4,zy4 = np.loadtxt('Insert.txt',delimiter=' ',unpack=True)
jx4,jy4 = np.loadtxt('Intercala.txt',delimiter=' ',unpack=True)

plt.plot(x4, y4, color="green", label="merges")
#plt.plot(sx4, sy4, color="black", label="mergedf")
plt.plot(zx4, zy4, color="red", label="insertion")
plt.plot(jx4, jy4, color="blue", label="Intercala")

plt.axis("on")
plt.xlabel("n")
plt.ylabel("Time(ms)")
plt.legend()
```
